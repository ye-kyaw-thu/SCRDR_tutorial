## SCRDR Tokenizer: Overall Performance Comparison

|Metric|Character-level|Syllable-level|  
|---|---|---|
|Word Precision|90.89%|92.85%|  
|Word Recall|84.41%|88.68%|  
|Word F1-Score|87.53%|**90.72%**|  
|Char/Unit Accuracy|96.17%|92.70%|  
|Char Macro F1|94.32%|92.63%|  

လုပ်ခဲ့တဲ့ Experiment က word segmentation or tokenization မို့လို့ Word-level F1 ကို အဓိက ထားကြည့်ကြရပါလိမ့်မယ်။ Word-level F1 က tokenization အလုပ်အတွက် gold standard ပါပဲ။ အထက်မှာ မြင်ရတဲ့အတိုင်းပဲ ရလဒ်အားဖြင့်က Word F1-Score မှာ syllable level က သာပါတယ်။
