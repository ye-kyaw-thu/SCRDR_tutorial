## SCRDR Tagging Experiment

ဒီ notebook ကတော့ SCRDR ကိုသုံးပြီး POS Tagging experiment ကို လက်တွေ့ run ပြထားတာ ဖြစ်ပါတယ်။  

ဒီမှာ သုံးပြထားတဲ့ scrdr_tagger.py code က အချိန်တော်တော်ပေးပြီး ရေးဖြစ်ခဲ့ပါတယ်။ အလုပ်အားတဲ့အချိန်မှာ ရေးလိုက် experiment လုပ်ကြည့်လိုက်နဲ့ စုစုပေါင်း သုံးလခန့်ကြာတယ်လို့ မှတ်မိနေပါတယ်။ POS tagging နဲ့ word tokenizer ကလည်း အခြေခံအားဖြင့် tagging လုပ်တာပဲ၊ အတူတူပဲလို့ နားလည်တာမို့ အခု ရှဲပေးထားသလိုမျိုး ပရိုဂရမ်နှစ်ပုဒ် မဟုတ်ပဲ ပရိုဂရမ် တစ်ပုဒ်ထဲ အနေနဲ့ ပေါင်းပြီး ရေးတာမျိုးလည်း လုပ်ကြည့်ဖြစ်ခဲ့ပါတယ်။ သို့သော်လည်း လက်တွေ့ function တွေကို ရေးတဲ့အခါမှာ တထပ်တည်းမတူတာမို့လို့ ကိုယ်ရေးထားတဲ့ ကုဒ်ကို ပြန်ကြည့်ပြီး ရှုပ်တယ်လို့၊ တခြားလိုက်ဖတ်တဲ့ သူတွေအတွက် ပိုခက်နိုင်တယ်လို့ ယူဆပြီး နောက်ဆုံးမှာ နှစ်ပုဒ် သပ်သပ်စီပဲ ခွဲရေးလိုက်ပါတယ်။  

Vibe coding ကိုလည်း Deepseek, ChatGPT, Gemini, Claude တွေနဲ့ လိုတဲ့အပိုင်းကို စမ်းရေးခိုင်းလိုက်၊ ရေးပြီးထွက်လာတဲ့ ကုဒ်ကို ဖတ်ကြည့်လိုက်၊ run ကြည့်လိုက်လုပ်ပြီးလည်း အမျိုးမျိုး လေ့လာဖြစ်ခဲ့ပါတယ်။ ဘာသွားတွေ့ရသလဲ ဆိုတော့ Ripple Down Rules နဲ့ ပတ်သက်တဲ့ ကုဒ်ကို အစအဆုံး အဆင်ပြေအောင် ရေးပေးနိုင်တာမျိုးကို လက်ရှိ LLM တွေအနေနဲ့က တော်တော်လေး ခက်ခဲတာကို သွားတွေ့ရပါတယ်။ LLM တွေထဲမှာတော့ SCRDR အတွက် Claude ကပဲ အကောင်းဆုံးလို့ ပြောရပါလိမ့်မယ်။ Module တွေအနေနဲ့ခွဲရေးလိုက် မြန်မာစာ myPOS ဒေတာကို သုံးပြီးစမ်းလိုက်၊ ရလဒ်က မကောင်းလို့ ပြန်ပြင်လိုက်၊ ရလဒ်က မဆိုးပေမဲ့လည်း run time က အရမ်းကြာနေလို့ debugging လုပ်လိုက်နဲ့ လက်ရှိဗားရှင်း ရောက်အောင်၊ ကိုယ်လိုချင်တဲ့ အတိုင်းအတာ တစ်ခုထိ ရောက်အောင် တော်တော် အချိန်ပေးလိုက်ရပါတယ်။ နောက်ဆုံးမှာတော့ Claude နဲ့ပဲ refine ပြန်လုပ်ထားတာပါ။

## Evaluation with evaluate.py Code

2017 တုန်းက တင်ထားခဲ့တဲ့ CICLING 2017 စာတမ်းမှာတုန်းက သုံးခဲ့တဲ့ evaluate.py ကိုပဲ သုံးပြီး evaluation လုပ်ကြည့်ပါမယ်။ အဲဒါမှ fair comparison ဖြစ်မှာမို့။ တစ်ခုရှိတာက လက်ရှိ training လုပ်ဖို့အတွက် သုံးထားတဲ့ myPOS corpus က version 3 မို့လို့ 2017 တုန်းက သုံးခဲ့တဲ့ myPOS corpus ထက် သုံးဆ ကျော်ကွာပါတယ်။  

```python
%pwd
```

```python
!python2.7 ./evaluate.py otest.hyp ./pos_data/otest.txt
```

CICLING စာတမ်းမှာတုန်းက CRF, HMM စတဲ့ ML method ၅မျိုးနဲ့ RDR ကို နှိုင်းယှဉ်ခဲ့ပါတယ်။ အဲဒီတုန်းက လုပ်ခဲ့တဲ့ experiment တွေရဲ့ ရလဒ်က အောက်ပါအတိုင်းပါ။ Open test ဒေတာနဲ့ ရတဲ့ RDR ရဲ့ ရလဒ်က 97.05% ပါ။    

|Methods|Closed|Test-set|Open Test-set|  
| ---| --- | --- | --- |
|CRFs|97.77%|95.05%|  
|HMM|97.31%|96.43%|  
|MaxEnt|96.55%|96.31%|  
|RDR|98.42%|**97.05%**|  
|SVM|**99.83%**|93.55%|  
|Two-Hours|95.83%|92.87%|  

အဲဒါကြောင့် လက်ရှိ scrdr_tagger.py နဲ့ myPOS (version 3.0) ဒေတာနဲ့ ရလဒ်ဖြစ်တဲ့ F-Measure:  0.998217998218 က ပြတ်ပြတ်သားသား သာပါတယ်။
